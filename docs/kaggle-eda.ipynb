# AI assistant popularity from Google search demand

A quick EDA of the [Cleanor Search Index](https://www.kaggle.com/datasets/cleanorlabs/cleanor-search-index): which AI assistants people actually search for, worldwide and by country, from Google Keyword Planner historical metrics.

Source & methodology: https://github.com/cleanor-app/search-index · DOI 10.5281/zenodo.21225547

```python
import pandas as pd, glob
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi']=120
def find(name):
    h=glob.glob(f"/kaggle/input/**/{name}", recursive=True); return h[0]
```

## 1. Worldwide ranking
ChatGPT dominates category search demand; Gemini is a distant second.

```python
s=pd.read_csv(find("popularity-ai-summary.csv"))
top=s[s.scope=="WW"].nlargest(8,"avg_monthly_searches")
ax=top.plot(x="brand",y="avg_monthly_searches",kind="barh",color="#0a7cff",legend=False)
ax.invert_yaxis();ax.set_xlabel("avg monthly searches");ax.set_ylabel("");ax.set_title("AI assistants by worldwide search demand")
plt.tight_layout();plt.show()
```

## 2. Who is growing (year-over-year)
Newer entrants show triple-digit YoY growth even as ChatGPT holds the volume.

```python
g=s[s.scope=="WW"].nlargest(8,"avg_monthly_searches").sort_values("yoy_pct")
ax=g.plot(x="brand",y="yoy_pct",kind="barh",color=["#ef4444" if v<0 else "#16a34a" for v in g.yoy_pct],legend=False)
ax.set_xlabel("year-over-year change (%)");ax.set_ylabel("");ax.set_title("AI assistant demand growth (YoY)")
plt.tight_layout();plt.show()
```

## 3. ChatGPT demand by country

```python
m=pd.read_csv(find("popularity-ai-monthly.csv"))
latest=m.year_month.max()
cg=m[(m.brand=="ChatGPT")&(m.year_month==latest)].nlargest(12,"searches")
ax=cg.plot(x="country",y="searches",kind="barh",color="#4f5bff",legend=False)
ax.invert_yaxis();ax.set_xlabel("monthly searches");ax.set_ylabel("");ax.set_title(f"ChatGPT search demand by country ({latest})")
plt.tight_layout();plt.show()
```
